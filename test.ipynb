```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def generateData(num_samples):
    np.random.seed(42)
    study_hours = np.random.randint(0, 11, size=num_samples)
    attendance = np.round(np.random.uniform(0.5, 1.0, size=num_samples), 2)
    grades = np.round(0.3 * study_hours + 4.5 * attendance, 1)
    data = np.column_stack((study_hours, attendance, grades))
    return data


num_samples = 200
data = generateData(num_samples)
```

```python
columns = ['study_hours', 'attendance', 'grade']
df = pd.DataFrame(data, columns=columns)
```

```python
print(df)
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(x='study_hours', y='attendance', data=df)
plt.title("Study Hours and Attendance Scatterplot")
plt.xlabel('Study Hours')
plt.ylabel('Attendance')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.regplot(x='study_hours', y='attendance', data=df)
plt.grid(True)
plt.show()
```
