```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import RandomForestRegressor
from sklearn import metrics
```

```python
gold_data = pd.read_csv('gld_price_data.csv')

gold_data
```

```python
gold_data.info()
gold_data.describe()
```

```python
gold_data_corr = gold_data.drop('Date', axis=1)
```

```python
correlation = gold_data_corr.corr()
```

```python
sns.heatmap(correlation, annot=True, cbar=True, fmt='.1f', cmap='Blues')
```

```python
print(correlation['GLD'])
```

```python
sns.distplot(gold_data['GLD'], color='green')
```

```python
X = gold_data.drop(['Date', 'GLD'], axis=1)
Y = gold_data['GLD']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
regressor = RandomForestRegressor(n_estimators=100)
```

```python
regressor.fit(X_train, Y_train)
```

```python
test_data_pred = regressor.predict(X_test)
```

```python
error = metrics.r2_score(Y_test, test_data_pred)
print(error)
```

```python
Y_test = list(Y_test)
```

```python
plt.plot(Y_test, color='blue', label='Actual value')
plt.plot(test_data_pred, color='red', label='Predicted value')
plt.xlabel('Number of values')
plt.ylabel('GLD price')
plt.legend()
```
